# Win-Draw-Lose Interactive

```python
# Create new feature for classifying outcome
def win_team(data):
    if data['home_team_goal'] > data['away_team_goal']:
        return data['home_team_name']
    elif data['home_team_goal'] < data['away_team_goal']:
        return data['away_team_name']
    elif data['home_team_goal'] == data['away_team_goal']:
        return 'Draw'

def lose_team(data):
    if data['home_team_goal'] > data['away_team_goal']:
        return data['away_team_name']
    elif data['home_team_goal'] < data['away_team_goal']:
        return data['home_team_name']
    elif data['home_team_goal'] == data['away_team_goal']:
        return 'Draw'
    
def game_outcome(data):
    if data['home_team_goal'] > data['away_team_goal']:
        return 'Home Win'
    elif data['home_team_goal'] < data['away_team_goal']:
        return 'Away Win'
    elif data['home_team_goal'] == data['away_team_goal']:
        return 'Draw'    
```

```python
team_2_df['win_team'] = team_2_df.apply(lambda x: win_team(x), axis=1)
team_2_df['game_outcome'] = team_2_df.apply(lambda x: game_outcome(x), axis=1)
team_2_df['lose_team'] = team_2_df.apply(lambda x: lose_team(x), axis=1)
```
