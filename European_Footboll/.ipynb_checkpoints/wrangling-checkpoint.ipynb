## What is the connector between three table above?
#### home_player_x

```python
def get_home_overall_rating(match, player):
    match_date = pd.to_datetime(match['date'])
    player_date = pd.to_datetime(player['date'])

    for i in range(1, 12):
        player = 'home_player_'+str(i)
        player_id = match[player]
        
        if np.isnan(player_id) == False:
            overall = player_df[(player_df['player_api_id'] == player_id)]
            current_overall = player[match_date > player_date]
            print(current_overall)
                                     
```

```python
att = match_df.apply(lambda x : get_home_overall_rating(x, player_df), axis=1)
```
