## Checkin_price

```python
checkin_price = df['checkin_price']
```

```python
checkin_price.describe(include='all')
```

```python
# Take only 99% of samples, for avoiding noise (max = 94157,..)
q = checkin_price.quantile(q=.99)
checkin_price_ = checkin_price[checkin_price <= q]
```

```python
# Clean 'checkin_price'
df = df[checkin_price <= q]
```

```python
checkin_price_.describe()
```

```python
bucket_table(series=checkin_price_, bins=5, plot=True)
```

```python
# 'chekin_price' by 'province'
df['checkin_price'].groupby(df['province']).mean().sort_values().plot.barh()
```

```python
# 'checkin_prince' by 'exp_type'
df['checkin_price'].groupby(df['exp_type']).mean().sort_values().plot.barh()
```

```python
# 'checkin_prince' by 'region'
df['checkin_price'].groupby(df['region']).mean().sort_values().plot.barh()
```

```python
# 'checkin_prince' by 'property_type'
df['checkin_price'].groupby(df['property_type']).mean().sort_values().plot.barh()
```

```python
# 'checkin_prince' by 'district'
df.groupby(['property_type', 'district'])[['checkin_price']].mean().sort_values(by='checkin_price', ascending=False)
```

```python
# 'checkin_price' by 'province' and 'property_type'
pro_pro_price = df.groupby(['province', 'property_type'])[['checkin_price']].mean()
# for i, j in zip(pro_pro_price.index, pro_pro_price.values):
#     print(i, ':', j)
pd.set_option('display.max_rows', pro_pro_price.shape[0]+1)
pro_pro_price
```

```python
# 'checkin_price' by 'property_type' and 'exp_type'
df.groupby(['exp_type', 'property_type'])[['checkin_price']].mean()
```

## Results:
* Overall mean price = 51.34

* 75% is less than 61.38/ over 80% focus on (3. - 75.99)

* Checkin price (average) by province:
    * Highest is BaRia-VungTau.
    * Top 5 is beach experience.
    
* Checkin price (average) by experience type:
    * Highest average price is Beach, then Urban and Moutain.
  
* Checkin price (average) by region:
    * Highest average price is North.
    * But the variability is pretty trivial.
  
* Checkin price (average) by property type:
    * Highest average price is Villa.
    * Top 5 by district is all Villa.
    * Villa is the most expensive property type in every province and exp_type.

## Price (nightly - weekly)

```python
# nightly_price
nightly_price = df['nightly_price']
```

```python
nightly_price.describe(include='all')
```

```python
# weekly_price
weekly_price = df['weekly_price']
```

```python
# take 99% of samples
q = weekly_price.quantile(q=.99)
# clean weekly_price
df = df[weekly_price <= q] 
```

```python
weekly_price.describe(include='all')
```

```python
price = df[['nightly_price', 'weekly_price']]
```

```python
price.plot.box()
```

```python
nightly_price = price['nightly_price']
weekly_price = price['weekly_price']
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

nightly_price.plot(ax=axes[0], kind='hist')
weekly_price.plot(ax=axes[1], kind='hist')
```

```python
n_pri_pro = nightly_price.groupby(df['province']).mean().sort_values()
w_pri_pro = weekly_price.groupby(df['province']).mean().sort_values()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1)
fig
n_pri_pro.plot(ax=axes[0], kind='barh')
w_pri_pro.plot(ax=axes[1], kind='barh')
```

```python
n_pri_exp = nightly_price.groupby(df['exp_type']).mean().sort_values()
w_pri_exp = weekly_price.groupby(df['exp_type']).mean().sort_values()

n_pri_reg = nightly_price.groupby(df['region']).mean().sort_values()
w_pri_reg = weekly_price.groupby(df['region']).mean().sort_values()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2)

n_pri_exp.plot(ax=axes[0, 0], kind='barh')
w_pri_exp.plot(ax=axes[0, 1], kind='barh')

n_pri_reg.plot(ax=axes[1, 0], kind='barh')
w_pri_reg.plot(ax=axes[1, 1], kind='barh')
```

```python
n_pri_per = nightly_price.groupby(df['property_type']).mean().sort_values()
w_pri_per = weekly_price.groupby(df['property_type']).mean().sort_values()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1)

n_pri_per.plot(ax=axes[0], kind='barh')
w_pri_per.plot(ax=axes[1], kind='barh')
```

```python
df.groupby(['property_type', 'district'])[['nightly_price', 'weekly_price']].mean().sort_values(by='weekly_price', ascending=False)
```

```python
pro_pro_price = df.groupby(['province', 'property_type'])[['nightly_price', 'weekly_price']].mean()
pd.set_option('display.max_rows', pro_pro_price.shape[0]+1)
pro_pro_price
```
