## Dealing with Missing Values

```python
for i, j in zip(df.isnull().sum().index, df.isnull().sum()):
    print('{}: {}'.format(i, j))
```

```python
data = df.drop(['lat', 'lon', 'host', 'address', 'popular', 'monthly_price', 'security_fee'], axis=1)
```

```python
def max_stay_boolean(data):
    if np.isnan(data):
        return 'no_limit'
    else:
        return 'limit'

data['max_stay'] = data['max_stay'].apply(max_stay_boolean)
```

## Distribution

### Describe_Before

```python
data[['num_bedrooms', 'num_bathrooms', 'num_beds', 'max_guests',
 'checkin_price', 'weekly_price', 
 'cleaning_fee','additional_guests_fee', 'additional_children_fee', 
 'additional_infants_fee', 'standard_guests', 'standard_children',
 'standard_infants', 'min_stay']].describe(include='all')
```

#### 'price'

```python
# Checkin_price / nightly
data = data[data['checkin_price'] <= data['checkin_price'].quantile(q=.99)]
# weekly_price
data = data[data['weekly_price'] <= data['weekly_price'].quantile(q=.99)]
```

#### 'fee'

```python
# Cleaning_fee
data = data[data['cleaning_fee'] <= data['cleaning_fee'].quantile(q=.99)]
# additional_guests_fee
data = data[data['additional_guests_fee'] <= data['additional_guests_fee'].quantile(q=.99)]
```

#### 'room_quality'

```python
# Number of bedrooms
data = data[data['num_bedrooms'] <= data['num_bedrooms'].quantile(q=.99)]
# Number of bathrooms
data = data[data['num_bathrooms'] <= data['num_bathrooms'].quantile(q=.99)]
# Number of beds
data = data[data['num_beds'] <= data['num_beds'].quantile(q=.99)]
# Max_guest
data = data[data['max_guests'] <= data['max_guests'].quantile(q=.99)]
# std_guests
data = data[data['standard_guests'] <= data['standard_guests'].quantile(q=.99)]
# std_children
data = data[data['standard_children'] <= data['standard_children'].quantile(q=.99)]
# min_stay
data = data[data['min_stay'] <= data['min_stay'].quantile(q=.99)]
```
