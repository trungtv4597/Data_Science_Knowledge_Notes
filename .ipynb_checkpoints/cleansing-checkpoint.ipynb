### Address


```python
add_f = df[['id', 'address']]
```

```python
def extract_city(data):
    container = data.split(',')
    try:
        return container[-2].replace(" ", "")
    except:
        return 'Unknown'
```

```python
add_f['city'] = add_f['address'].apply(extract_city)
```

```python
def extract_district(data):
    container = data.split(',')
    try:
        return container[-3].replace(" ", "")
    except:
        return 'Unknown'
```

```python
add_f['region'] = add_f['address'].apply(extract_district)
```
